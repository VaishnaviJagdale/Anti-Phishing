```python
#Univariate Analysis
def discrete_univariate(df,feature):
    total=len(df[feature])
    count=df[feature].nunique()
    plt.figure(figsize=(count+1,5))
    ax=sns.countplot(data=df,x=feature,order=df[feature].value_counts().index.sort_values(ascending=False))
    for a in ax.patches:
        label="{:.1f}%".format(100*a.get_height()/total)
        x=a.get_x()+a.get_width()/2
        y=a.get_height()
        ax.annotate(label,(x,y),ha='center',va='center',xytext=(0,5),size=8,textcoords="offset points")
    plt.show()
```

```python
discrete_univariate(df,'CLASS_LABEL')
```

```python
#Bivariate Analysis
def bivariate_analysis(df, feature_list):
    plt.figure(figsize=(15,10))
    for i, variable in enumerate(feature_list):
        plt.subplot(3,2, i+1)
        sns.boxplot(df['CLASS_LABEL'], df[variable], palette='viridis')
        plt.tight_layout()
        plt.title('Class Label vs '+variable)
    plt.show()
    
```

```python
def heatmap(df,start,end):
    new_df=pd.DataFrame(df.iloc[:,start:end])
    plt.figure(figsize=(10,8))
    sns.heatmap(new_df.corr(),annot=True,fmt='.3f')
    
```

# Splitting Dataset into training and testing dataset(training set=80% and testing set=20%)



```python
x=df.drop(columns=['id','CLASS_LABEL'],axis=1)
y=df['CLASS_LABEL']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

# XGBoost

```python
#Model fitting
from xgboost import XGBClassifier
clf=XGBClassifier()
clf.fit(x_train,y_train)
```

```python
#prediction for testing values
y_pred=clf.predict(x_test)
y_pred
```

```python
#Prediction for training values
y_train_pred=clf.predict(x_train)
y_train_pred
```

```python
#Performance Evaluation
from sklearn.metrics import mean_absolute_error,mean_squared_error,recall_score,precision_score,accuracy_score
```

```python
#MAE
mean_absolute_error(y_test,y_pred)
```

```python
#MSE
mean_squared_error(y_test,y_pred)
```

```python
#RMSE
mean_squared_error(y_test,y_pred,squared=False)
```

```python
#Recall
recall_score(y_test,y_pred)
```

```python
#precision
precision_score(y_test,y_pred)
```

```python
#accuracy
accuracy_score(y_test,y_pred)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
